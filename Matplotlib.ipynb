# Matplotlib Examples

### Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(1,10,10)
y = np.random.randint(1,10,10)

plt.scatter(x,y)
plt.tight_layout()
plt.show()
```

```python
x = np.random.randint(1,10,10)
y = np.random.randint(1,10,10)

plt.figure(figsize = (7,5))
plt.style.use("fivethirtyeight") #adding style. Check various styles on matplotlib page

plt.scatter(x,y)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")

plt.tight_layout()
plt.show()
```

```python
x = np.random.randint(1,10,10)
y = np.random.randint(1,10,10)
size = np.random.randint(100,300,10)
color = np.random.randint(1,10,10)

plt.figure(figsize = (7,5))
plt.style.use("default")

plt.scatter(x,y,s = size,c = color, edgecolor = 'black', alpha = 0.75) #adding size, color, edgecolor, alpha

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")

plt.tight_layout()
plt.show()
```

```python
x = np.random.randint(1,10,10)
y = np.random.randint(1,10,10)
size = np.random.randint(100,300,10)
color = np.random.randint(1,10,10)

plt.figure(figsize = (7,5))
plt.style.use("default")

plt.scatter(x,y,s = size, c = color, cmap = 'Greens' ) #adding cmap and color bar. cmap works with color parameter
cbar = plt.colorbar()
cbar.set_label("Satisfaction")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")

plt.tight_layout()
plt.show()
```
