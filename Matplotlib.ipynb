## 1. Line Chart

- A line plot is a graph that shows data along a number line. 
- It is best to use a line plot when comparing small changes in data. It is a quick, simple way to organize and represent your data.

__Plotting Line chart__

```python

x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)

plt.style.use("grayscale")

plt.figure(figsize=(7,5))

plt.plot(x,y1,"b-x", label = '$y1 = x^2$') #'b' stands for blue colour and '-' is used for line type and 'x' is used for marker
plt.plot(x,y2,"r-", label = '$y2 = x^3$', lw = 3) #lw stands for linewidth. We can also write linewidth=

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.legend(fontsize = 14, bbox_to_anchor = (1,1)) #adding legend options. bbox_to_anchor option anchors the legend outside the axes

plt.tight_layout()
plt.show()
```

### 2. Bar Plot

- Bar graphs are used to compare things between different groups or to track changes over time. However, when trying to measure change over time, bar graphs are best when the changes are larger.
- A good practice is to always sort the bar graphs

__Simple bar chart__

```python
x1 = [1,2,3,4]
y1 = [5,7,9,6]

plt.figure(figsize=(8,6))

plt.gcf().set_facecolor("pink")
plt.bar(x1,y1, label = 'First', color = '#42B300')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar Chart")

plt.legend()
plt.show()
```

#### When the lables are too long to appear clearly on x-axis, then Horizontal bar graph is the best way to present that data.

__Horizontal bars using barh function__

```python
x1 = [1,2,3,4]
y1 = [5,7,9,6]
error = [1,2,2,1]

plt.figure(figsize=(8,6))

plt.gcf().set_facecolor("pink")
plt.barh(x1,y1, label = 'First', xerr = error, color = 'blue') 

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar Chart")

plt.legend()
plt.show()
```

__Plotting two bars side by side using x1 and x2 consecutively__


```python
x1 = [1,3,5,7]
x2 = [2,4,6,8]
y1 = [7,8,5,10]
y2 = [17,18,29,40]

plt.figure(figsize=(8,6))

plt.bar(x1,y1, label = 'First', color = '#42B315')
plt.bar(x2,y2, label = 'Second', color = '#94E413')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar Chart")

plt.legend()
plt.show()
```

__Displaying values at the top of vertical bars__

```python
num1 = np.arange(1,10,2)
num2 = np.arange(2,11,2)

plt.figure(figsize = (10,6))

plt.bar(num1, num1**2, width = 0.3, color = '#FF6F00')
plt.bar(num2, num2**2, width = 0.3, color = '#FFB300')

for x,y in zip(num1,num1**2):
  plt.text(x, y+0.05, '%d'%y, ha='center', va = 'bottom')

for x,y in zip(num2,num2**2):
  plt.text(x,y+0.5, '%d'%y, ha = 'center', va = 'bottom')

plt.show()
```

__Three bars with text on top of them__

```python
plt.figure(figsize=(10,7))
plt.style.use("default")

barWidth = 0.25

y1 = np.array([17,18,29,40])
y2 = np.array([20,21,22,23])
y3 = np.array([5,9,11,12])

pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]

plt.bar(pos1, y1, color = '#E0BBE4', width=barWidth, label = 'Open Tickets', edgecolor = 'black')
plt.bar(pos2, y2, color = '#FEC8D8', width=barWidth, label = 'Closed Tickets', edgecolor = 'black')
plt.bar(pos3, y3, color = '#9EE09E', width=barWidth, label = 'Cancelled Tickets', edgecolor = 'black')

plt.xticks([i+barWidth for i in range(len(y1))],['Amit', 'Sumit', 'Jimit', 'Vinit']) # used for x ticks where the name appears below the middle bar

for x,y in zip(pos1, y1):
  plt.text(x,y+0.5, '%d' %y, ha = 'center', va = 'bottom')

for x,y in zip(pos2, y2):
  plt.text(x,y+0.5, '%d' %y, ha = 'center', va = 'bottom')

for x,y in zip(pos3, y3):
  plt.text(x,y+0.5, '%d' %y, ha = 'center', va = 'bottom')

plt.gca().set(title = 'Bar Chart', ylabel = 'Frequency')  # gca() is used to getting axes and performing functions on axes instead of object declaration.

plt.legend(loc = 'upper left')


plt.show()
```

__Stacked bars and text inside them__

```python
x1 = ['Amit','Sumit','Jimit','Vinit']
y1 = np.array([17,18,29,40])
y2 = np.array([20,21,22,23])
y3 = np.array([5,9,11,12])


barWidth = 0.25

plt.figure(figsize=(7,7))
plt.style.use("default")

plt.bar(x1,y1,label = 'Open Tickets', width = 0.5, color = '#CC99C9', edgecolor = 'black')
plt.bar(x1,y2,label = 'Closed Tickets', bottom =  y1, width = 0.5, color = '#9EC1CF', edgecolor = 'black') #use bottom for stacking the bar on top of previous bar
plt.bar(x1,y3,label = 'Cancelled Tickets', bottom = y1+y2, width = 0.5, color = '#9EE09E', edgecolor = 'black')

#plt.text is used to plot the text in the figure at the specific location
for x,y in zip(x1,y1):
  plt.text(x,(y/2), '%d'%y, ha = 'center', va = 'bottom')
for x,y,z in zip(x1,y2,y1):
  plt.text(x,(y/2)+z, '%d'%y, ha = 'center', va = 'bottom')
for x,y,z,a in zip(x1,y3,y2,y1):
  plt.text(x,(y/2)+z+a-1, '%d'%y, ha = 'center', va = 'bottom')


plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bar & Chart')

plt.legend()
plt.show()
```

### 3. Pie Chart 

- A pie chart is used when we want to present the part-to-whole ratio in the data
- It is generally used to represent the categorical data. But too many categories can clutter the chart and can make it difficult to read. Do not use pie chart if you have more than 5 categories or sections. Use stacked bar plots in that case.
- Try to place the largest slice at the top (12 o clock) moving clockwise
- Avoid comparing one pie chart with another
- Avoid using 3-D pie charts

__Simple pie chart__

```python
plt.figure(figsize=(7,5))
plt.style.use('default')

area = [48, 30, 20, 15]

labels = ['Low', 'Medium','High','Critical']
colors = ['#DC828F', '#F7CE76','#E8D6CF','#8C7386']

plt.pie(area, labels = labels, colors = colors)

plt.show()
```

__Move portion of the pie out__

```python
plt.figure(figsize=(7,5))
plt.style.use('bmh')

area = [48, 30, 20, 15]

labels = ['Low', 'Medium','High','Critical']
colors = ['#DC828F', '#F7CE76','#E8D6CF','#8C7386']

#Here 0.1 means 10 percent. ie the third piece will move out 10 percent
explode = [0,0,0.1,0]

plt.pie(area, labels = labels, colors = colors, explode = explode)
plt.tight_layout()

plt.show()
```

__Portions of the pie are called wedges.<br>
Pie chart using wedgeprops option__

```python
plt.figure(figsize=(7,5))
plt.style.use('bmh')

labels = ['India', 'Canada', 'Japan', 'Australia', 'Russia']
sizes = [31, 19, 15, 14, 21]  # Add upto 100%
colors = ['#DC828F', '#F7CE76','#E8D6CF','#8C7386', 'Red']

#autopct option is used to show the percentage share of each portion
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors,
        wedgeprops={"edgecolor":"black",'linewidth': 1, 'linestyle': 'solid'})
 
# Equal aspect ratio ensures that pie is drawn at the center of the figure.
plt.axis('equal')
plt.tight_layout() 
plt.show()
```

### 4. Histograms 

- Histograms are used to plot quantitative data with ranges. The data is grouped into bins or intervals.
- The are different from bar charts. Bar charts are used to plot categorical data.

__Simple histogram figure using 40 number of bins__

```python
plt.figure(figsize=(7,5))
x = np.random.normal(size = 2000) # Generates 2000 standard normal random numbers
plt.hist(x, bins = 40, color = 'yellowgreen', edgecolor = 'black')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')

plt.show()
```

__We can define our own bins as well__

```python
id = np.arange(1,101)
age = np.random.normal(1,100,100)
bins = [0,20,40,60,80,100]

plt.hist(age, bins = bins, edgecolor = 'black')

plt.title('Age distribution')
plt.xlabel('Ages')

plt.tight_layout()

plt.show()
```

__- If data has some outliers and if we plot the histogram, the bin may not be visible sometimes.<br> We can overcome the situation using logarithmic scales__

__Data with outliers plotted where the last bin data is negligible__

```python
age = np.random.randint(1,99,10000)
age = np.append(age,[110,111])
bins = [0,20,40,60,80,100,120]

plt.hist(age, bins = bins, edgecolor = 'black')

plt.title('Age distribution')
plt.xlabel('Ages')

plt.tight_layout()

plt.show()
```

__Now plotting the data using logarithmic scales with log = True Option__

```python
age = np.random.randint(1,99,10000)
age = np.append(age,[110,111])
bins = [0,20,40,60,80,100,120]

plt.hist(age, bins = bins, edgecolor = 'black', log = True)

plt.title('Age distribution')
plt.xlabel('Ages')

plt.tight_layout()

plt.show()
```

### 5. Scatter Plots

- Scatter plot is used when you have two variables that pair well together. 
- Scatter diagram is used to show their relationship between the two variables and see if the variables are correlated or not. If yes, then positively correlated or negatively correlated.

__Simple Scatter Plot__

```python
x = np.random.randint(1,10,10)
y = np.random.randint(1,10,10)

plt.figure(figsize = (7,5))
plt.style.use("fivethirtyeight") #adding style. Check various styles on matplotlib page

plt.scatter(x,y)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")

plt.tight_layout()
plt.show()
```

__Adding other options of scatter plot__


```python
x = np.random.randint(1,10,10)
y = np.random.randint(1,10,10)
size = np.random.randint(100,300,10)
color = np.random.randint(1,10,10)

plt.figure(figsize = (7,5))
plt.style.use("default")
plt.gcf().set_facecolor("white")

plt.scatter(x,y,s = size,c = color, edgecolor = 'black', alpha = 0.75) #adding size, color, edgecolor, alpha

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")

plt.tight_layout()
plt.show()
```

__Adding cmap and color bar options__

```python
x = np.random.randint(1,10,10)
y = np.random.randint(1,10,10)
size = np.random.randint(100,300,10)
color = np.random.randint(1,10,10)

plt.figure(figsize = (7,5))
plt.style.use("default")

plt.scatter(x,y,s = size, c = color, cmap = 'RdYlBu' ) #adding cmap and color bar. cmap works with color parameter

cbar = plt.colorbar()
cbar.set_label("Satisfaction")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")

plt.tight_layout()
plt.show()
```

### 6. Area Plot 

- Area charts are used to show how values develop over time.
- Stacked area charts can be used to show multiple series with part-to-whole ratio/share
- There are 2 ways to build an area chart - using fill_between() function and using stackplot() function

__Simple Area chart__

```python
x = np.arange(1,31)
y = np.random.randint(2,11,size = 30)

plt.figure(figsize = (7,5))

plt.fill_between(x,y,color = '#E2BEF1')
plt.plot(x,y, color = 'black')

plt.title('Area Chart', fontsize = 16)

plt.show()
```

__Stacked area chart using a 2 dimensional (3X5) array__

```python
x=np.arange(1,6)
y=[[1,5,6,8,9], [2,3,7,10,12], [2,8,5,11,6]]
 
plt.stackplot(x,y, labels=['A','B','C'])
plt.legend(loc='upper left')
plt.show()
```

__Stacked area chart using 3 different lists for y axis__

```python
x=range(1,6)
y1=[1,3,6,7,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]
 
plt.stackplot(x,y1, y2, y3, labels=['A','B','C'])
plt.legend(loc='upper left')
plt.show()
```

### 7. Time series plot


Plotting time seried data is lot like line plot. But here we are using dates. And can perform various date functions like chanding the date format, etc

__Simple Time series plot__

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates

dates = [
         datetime(2021,5,21),
         datetime(2021,5,22),
         datetime(2021,5,23),
         datetime(2021,5,24),
         datetime(2021,5,25),
         datetime(2021,5,26),
         datetime(2021,5,27)
]

y = np.random.randint(1,10,7)

plt.plot_date(dates,y, ls = 'solid')

plt.gcf().autofmt_xdate() # autofmt_dates auto formats the alignment of dates by rotating the dates to a certain angle.
date_format = mpl_dates.DateFormatter('%b, %d %Y') # changin date format

plt.gca().xaxis.set_major_formatter(date_format) #using gca function to get axis object functionalities

plt.tight_layout()
plt.show()
```

### 8. Subplots

- So far we have been using an object of pyplot module as plt and its functions and attributes. But we did not create any object. We can use more object oriented functionalities and create an object using plt.figure and plt.subplots methods. 
- Subplots gives us much more flexibility and using subplot objects, we can do all things that we did using pyplot object.


*__Please understand figure is container holding all our plots (like a window) and the axes are the plots. While working with multiple figures and axes, we have to use the subplots method__*

```python
A = list(range(1,11))
B = np.power(A,2)
C = np.power(A,3)

x = np.arange(1,11)

fig = plt.figure() # create a figure object

ax = fig.add_subplots() # adds axes to the figure. By default subplots have (1,1) values

ax.plot(x,A, label = 'A')
ax.plot(x,B, label = 'B')
ax.plot(x,C, label = 'C')

#instead of using title, xlabel, and ylabel functions, with axes, we us set_title, set_xlabel, and set_ylabel functions.
ax.set_title("Subplotting")
ax.set_xlabel("X")
ax.set_ylabel("A/B/C")

ax.legend()

plt.tight_layout()

plt.show()
```

__Multiple axes by adding more axes. By default it is (1,1)__

```python
A = list(range(1,11))
B = np.power(A,2)
C = np.power(A,3)

x = np.arange(1,11)

fig,(ax1,ax2) = plt.subplots(2,1) # this function directly creates axes and attaches them to fig object

ax1.plot(x,A, label = 'A')
ax1.plot(x,B, label = 'B')
ax2.plot(x,C, label = 'C')

ax1.set_title("Subplotting 1")
ax1.set_xlabel("X")
ax1.set_ylabel("A/B")

ax2.set_title("Subplotting 2")
ax2.set_xlabel("X")
ax2.set_ylabel("C")

ax1.legend()
ax2.legend()

plt.tight_layout()

plt.show()
```

__Plotting multiple figures removing x labels, adjusting x_ticks only for bottom plot.__

```python
A = list(range(1,11))
B = np.power(A,2)
C = np.power(A,3)

x = np.arange(1,11)

fig,(ax1,ax2) = plt.subplots(2,1, sharex = True) # sharex option helps us in sharing the x ticks 

ax1.plot(x,A, label = 'A')
ax1.plot(x,B, label = 'B')
ax2.plot(x,C, label = 'C')

ax1.set_title("Subplotting 1")
ax1.set_ylabel("A/B")

ax2.set_xlabel("X")
ax2.set_ylabel("C")

ax1.legend()
ax2.legend()

plt.tight_layout()

plt.show()
```

__Plotting two different figures (two windows)__

```python
A = list(range(1,11))
B = np.power(A,2)
C = np.power(A,3)

x = np.arange(1,11)

fig1,ax1 = plt.subplots()
fig2,ax2 = plt.subplots()

ax1.plot(x,A, label = 'A')
ax1.plot(x,B, label = 'B')
ax2.plot(x,C, label = 'C')

#instead of using title, xlabel, and ylabel functions, with axes, we us set_title, set_xlabel, and set_ylabel functions.
ax1.set_title("Subplotting 1")
ax1.set_ylabel("A/B")

ax2.set_xlabel("X")
ax2.set_ylabel("C")

ax1.legend()
ax2.legend()

plt.tight_layout()

plt.show()
```
